```python
# Created an array of size [10,50] and used the random function to fill it with integers between 1 and 500 (inclusive)
array_new = np.random.randint(1,501, size = (10,50))
```

```python
# Loading above array to a DataFrame.
d_frame = pd.DataFrame(array_new)
d_frame.head(5)
```

#### 2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
# Calculating the sum of each row
sum_row = d_frame.sum(axis = 0)
sum_row
```

```python
# the sum of each column
sum_column = d_frame.sum(axis = 1)
sum_column
```

```python
# Sum of all columns
df_sum= d_frame.values.sum()
df_sum
```

#### 3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
# minimum; maximum; average; and standard deviation of all entries.
df_min = d_frame.values.min()
df_max = d_frame.values.max()
df_avg = d_frame.values.mean()
df_std = d_frame.values.std()
print("Minimum value of the above dataframe is:", df_min)
print("Maximum value of the above dataframe is:", df_max)
print("Average value of the above dataframe is:", df_avg)
print("Standard deviation of the above dataframe is:", df_std)
```

#### 4. Sort the DataFrame: a) sort on column 2; b) sort on row 2.


```python
# sorting column 2
d_frame.sort_values(by = 2,axis = 1, ascending = False)
```

```python
# sorting row 2
d_frame.sort_values(by = 2,axis = 0, ascending = False)
```
